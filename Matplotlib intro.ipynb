# Matplotlib example workflow

```python
# 1. prepare data
x = [11, 7, 24, 18]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(6, 6))

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. save and show
fig.savefig('simple-sample-plot.png')
```

# Pyplot vs matplotlib OO method

* Když rychle = pyplot method (.plot())
* Když složitě = OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim([45, 80])
```

```python
# OO method from the scratch
fix, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol');

# Add legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a mean line
ax.axhline(np.mean(over_50['chol']), 
          linestyle='--');
```

```python
# Subplot of thalach, chol, age

fix, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=heart_disease['age'],
                    y=heart_disease['thalach'],
                    c=heart_disease['target']);

# Customize the plot
ax.set(title='Heart Disease and Heart Rate Levels',
      xlabel='Age',
      ylabel='Heart Rate');

# Add legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a mean line
ax.axhline(np.mean(heart_disease['thalach']), 
          linestyle='--');
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3,
                              ncols=1,
                              figsize=(10, 15),
                              sharex=True);


# Setting first subplot
scatter = ax0.scatter(x=heart_disease['age'],
                     y=heart_disease['chol'],
                     c=heart_disease['target'])

ax0.set(title='Heart Disease and Heart Rate Levels',
       xlabel='Age',
       ylabel='Cholesterol')

ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(y=heart_disease['chol'].mean(), 
          linestyle='--');


# Setting second subplot
scatter = ax1.scatter(x=heart_disease['age'],
                     y=heart_disease['thalach'],
                     c=heart_disease['target'])

ax1.set(title='Heart Disease and Heart Rate Levels',
       xlabel='Age',
       ylabel='Heart Rate')

ax1.legend(*scatter.legend_elements(), title='Target')

ax1.axhline(y=heart_disease['thalach'].mean(), 
          linestyle='--');


# Setting third subplot
scatter = ax2.scatter(x=heart_disease['age'],
                    y=heart_disease['trestbps'],
                    c=heart_disease['target']);

ax2.set(title='Heart Disease and Trestbps Levels',
      xlabel='Age',
      ylabel='Bps');

ax2.legend(*scatter.legend_elements(), title='Target');

ax2.axhline(np.mean(heart_disease['trestbps']), 
          linestyle='--');

# Add Title to the entire figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

# OO method from the scratch
fix, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter');

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol');

# Add legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a mean line
ax.axhline(np.mean(over_50['chol']), 
          linestyle='--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True);


# Setting first subplot
scatter = ax0.scatter(x=heart_disease['age'],
                     y=heart_disease['chol'],
                     c=heart_disease['target'],
                      cmap='winter')

ax0.set_xlim([30, 80])
ax0.set_ylim([100, 450])

ax0.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(y=heart_disease['chol'].mean(), 
          linestyle='--');


# Setting second subplot
scatter = ax1.scatter(x=heart_disease['age'],
                     y=heart_disease['thalach'],
                     c=heart_disease['target'],
                     cmap='winter')

ax1.set_xlim([30, 80])
ax1.set_ylim([80, 200])

ax1.set(title='Heart Disease and Heart Rate Levels',
       xlabel='Age',
       ylabel='Heart Rate')

ax1.legend(*scatter.legend_elements(), title='Target')

ax1.axhline(y=heart_disease['thalach'].mean(), 
          linestyle='--');


# Add Title to the entire figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
